## Visualizing Data with Charts

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from pandas import DataFrame, Series
```

```python
# set chart defaults
plt.rc('figure', figsize=(10,6))
```

```python
data = np.arange(10)
data
```

```python
plt.plot(data)
```

```python
# sometimes we want several charts in one figure
fig = plt.figure()
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
ax4 = fig.add_subplot(2,2,4)
```

```python
# format the appearance of the chart
plt.plot(np.random.randn(50).cumsum(), 'k--')
```

```python
_ = ax1.hist(np.random.randn(100), bins=20, color='k', alpha=0.3)
ax2.scatter(np.arange(30), np.arange(30) + 3 * np.random.randn(30))
_  # _ is just a simple variable name!!!
```

```python
ig, axes = plt.subplots(2,2, sharex=True, sharey=True)
for i in range(2):
    for j in range(2):
        axes[i,j].hist(np.random.randn(100), bins=20, color='k', alpha=0.3)
plt.subplots_adjust(wspace=0, hspace=0)
```
